1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
pivot_df = marketing_df.pivot_table(index='sales_channel', values=['customer_lifetime_value'], aggfunc='sum')
```

```python
pivot_df['customer_lifetime_value'] = pivot_df['customer_lifetime_value'].round(2)
pivot_df.sort_values(by='customer_lifetime_value', ascending=False)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
avg_pivot_df = marketing_df.pivot_table(index='gender', columns=['education'], values=['customer_lifetime_value'], aggfunc='mean')
avg_pivot_df = avg_pivot_df.round(2)
avg_pivot_df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
avg_pivot_df.columns.values
```

```python
# Flatten the MultiIndex columns and create proper melted dataframe
avg_pivot_df.columns = [col[0] if col[1] == '' else col[1] for col in avg_pivot_df.columns]
avg_pivot_df = avg_pivot_df.reset_index()
avg_pivot_df
```

```python
melted_df = pd.melt(avg_pivot_df, id_vars=['gender'], var_name='education', value_name='customer_lifetime_value')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(
    data=melted_df,
    x="education",
    y="customer_lifetime_value",
    hue="gender"
)
plt.title("Customer Lifetime Value by Education and Gender")
plt.xticks(rotation=70)
plt.ylabel("Customer Lifetime Value")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.catplot(
    data=melted_df,
    x="education",                         # Categorical variable for X-axis
    y="customer_lifetime_value",           # Numeric variable for Y-axis
    hue="gender",                          # Optional: for group comparison (side-by-side groups)
    kind="bar",                             # Type of plot: "bar", "box", "violin", etc.
    aspect=2
)
plt.title("Customer Lifetime Value by Education and Gender")
plt.xticks(rotation=85)
plt.ylabel("Customer Lifetime Value")
plt.show()
```
